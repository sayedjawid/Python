```python
df_clean = df.dropna(subset=['studie_timmar', 'provpoäng']).reset_index(drop=True)
df_clean
```

```python
sammanfattning = (
                    df.groupby('klass')['provpoäng']
                    .agg(antal = 'count', medel='mean', median='median', std='std')
                    .reset_index())

sammanfattning
```

```python
df.groupby('klass').agg(
    provpoäng_medel = ('provpoäng', 'mean'),
    provpoäng_std = ('provpoäng', 'std'),
    studie_timmar_median = ('studie_timmar', 'median') 

).reset_index()

# nytt_kolumnnanm = (kolumn, funktion)
```
