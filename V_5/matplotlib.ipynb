```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 



df = pd.read_csv('resultat_städad.csv')
summary_klass = pd.read_csv('summary_by_class.csv')

df.head()
```

```python
summary_klass.head()
```

```python
plt.figure(figsize=(5,5)), plt.scatter(df.studietimmar, df['provpoäng']), plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5, color='green')
plt.title('Studietimmar vs provpoäng')
plt.xlabel('Studietimmar')
plt.ylabel('Provpoäng')
```

```python
fig, ax = plt.subplots()

df.provpoäng.plot(kind='hist', bins=10, edgecolor='red' , ax = ax, alpha=0.6, color='Green')
ax.set_title('Fördelning av provpoäng')
ax.set_xlabel('Poäng')
ax.set_ylabel('Antal')

ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5, color='blue', axis='y')
```

```python
fig, ax = plt.subplots()
x = [1,2,3]
y1 = [1,2,1]
y2 = [0.5,1.5,1]

ax.plot(x, y1, marker='o', linewidth=2, label='Linje 1')
ax.plot(x, y2, marker='s', linewidth=2, label='Linje 2')
ax.set_title('Axes-mönster')
ax.set_xlabel('X-axel')
ax.set_ylabel('Y-axel')
ax.grid(True)
ax.legend()  
```

```python
fig, axes = plt.subplots(2,2, figsize=(10, 8), sharex=False, sharey=False)

axes[0,0].hist(df.provpoäng.dropna(), bins=10, color='skyblue', edgecolor='black')
axes[0,0].set_title('Fördelning av provpoäng')
axes[0,0].set_xlabel('Poäng'); axes[0,0].set_ylabel('Antal'); axes[0,0].grid(True, axis='y')


axes[0,1].scatter(df.studietimmar, df.provpoäng, color='orange', alpha=0.6)
axes[0,1].set_title('Studietimmar vs provpoäng')
axes[0,1].set_xlabel('Studietimmar'); axes[0,0].set_ylabel('Provpoäng'); axes[0,0].grid(True)



axes[1,0].bar(summary_klass.klass, summary_klass.medel_poäng, color='orange', alpha=0.6)
axes[1,0].set_title('Medelpoäng per klass')
axes[1,0].set_xlabel('Klass'); axes[1,0].set_ylabel('Medel'); axes[1,0].grid(True, axis='y')

axes[1,1].axis('off')  # Dölj den tomma subploten

fig.suptitle('Äversokt: Tre sätt att visa data', fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.9)
```
